```python
import pandas as pd
from datetime import datetime
from dateutil import parser
```

```python
df = pd.read_excel(r"C:\Users\Lessard_M\Documents\Python\tor_fileuploader\data source\test_crstemplate_GG.xlsx",header = 1)
df = df.loc[1:].reset_index(drop=True)
```

```python
df["Commitment date (dd-mm-yyyy)"] = pd.to_datetime(df["Commitment date (dd-mm-yyyy)"],format = "%d-%m-%Y",dayfirst = True) 
```

```python
def date_converter (df,colname,output_format = "%d-%m-%Y"):
    """Converts dates to dd-mm-yyyy format. Input dataframe and column to be converted. The default output_format is "dd-mm-yyyy" """
    def save_converter(x):
        if pd.isna(colname):
            return pd.NaT
        try:
            return pd.to_datetime(x, format = output_format,dayfirst = True,error = "coarse")
        except: 
            pass
        try:
            return parser.parse(str(x),dayfirst = True,fuzzy = True)
        except:
            return pd.NaT
    df[colname] = df[colname].apply(save_converter)
    df[colname] = df[colname].dt.strftime(output_format)
    return df
```

```python
df_new = date_converter(df,"Commitment date (dd-mm-yyyy)","%d-%m-%Y")
```
